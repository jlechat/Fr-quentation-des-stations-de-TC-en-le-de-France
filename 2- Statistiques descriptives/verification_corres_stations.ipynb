## Identification des stations avec plusieurs lda
On cherche les stations pour lesquelles il y a plusieurs lda pour une même station ou un même lda pour deux stations

```python
unique_lda_check = data.groupby('LIBELLE_ARRET')['lda'].nunique()

# Afficher les LIBELLE_ARRET avec plus d'un lda
non_unique_lda = unique_lda_check[unique_lda_check > 1]
print(non_unique_lda)
```

```python
unique_lda_check = data.groupby('lda')['LIBELLE_ARRET'].nunique()

# Afficher les LIBELLE_ARRET avec plus d'un lda
non_unique_lda = unique_lda_check[unique_lda_check > 1]
print(non_unique_lda)
```
